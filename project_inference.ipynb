```python
api_key = userdata.get('OPENAI_API_KEY')

client = OpenAI(api_key = api_key)
```

```python
def evaluate_solution(problem, candidate_solution):

    prompt = f"""
    You are a math expert reviewing the logical correctness of a step-by-step solution to a math problem.

    ### Math Problem:
    {problem}

    ### Candidate Solution:
    {candidate_solution}

    Your task is to examine each line of the Candidate Solution *individually*, checking whether it logically follows from the previous steps. If a line contains incorrect or flawed reasoning, revise **only that line** to make it logically consistent, without changing the overall structure or intent of the solution.

    Do not rewrite the entire solution — only fix the lines that are incorrect or illogical, and leave the correct lines unchanged.

    Present the final result as the corrected step-by-step solution, preserving the original format.

    ### Your revised version:
    """

    messages = [
        {"role": "system", "content": "You are a helpful and accurate math evaluator."},
        {"role": "user", "content": prompt}
    ]

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",  # or "gpt-4"
        messages=messages,
        temperature=0.0,  # for deterministic judgment
    )

    result = response.choices[0].message.content.strip()

    return result
```

```python
def transfer_prompt(user_input):
  prompt= f"""You are a math assistant that translates Vietnamese math problem descriptions into formal English.

The user will input a math problem in Vietnamese, possibly including mathematical expressions written in plaintext (e.g., "x^2 + y = 3").

Your tasks are:
1. Translate the entire problem into clear and formal English suitable for a math textbook or exam.
2. Convert all mathematical expressions into LaTeX format.
3. Do NOT solve the problem — only translate and format it.

Output the result as a Markdown-formatted string using LaTeX where appropriate.

### Example:

Input (Vietnamese):
"Giải phương trình bậc hai: x^2 - 5x + 6 = 0"

Output (English):
Solve the quadratic equation:


Now, here is the user input to translate:

### User input:
{user_input}

### Your translate version:
"""
  messages = [
      {"role": "system", "content": "You are a math assistant that translates Vietnamese math problem descriptions into formal English."},
      {"role": "user", "content": prompt}
  ]

  response = client.chat.completions.create(
      model="gpt-3.5-turbo",  # or "gpt-4"
      messages=messages,
      temperature=0.0,  # for deterministic judgment
  )

  result = response.choices[0].message.content.strip()

  return result
```

```python
def read_from_image(image_path):
    # Read image as Base64
    with open(image_path, "rb") as f:
        image_bytes = f.read()
        base64_image = base64.b64encode(image_bytes).decode("utf-8")

    # Send API request to GPT-4 Vision
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": "You are a math assistant that translates Vietnamese math problem descriptions into formal English. Convert math expressions into LaTeX but do NOT solve them."
            },
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": "Please read this math problem and translate it."},
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/png;base64,{base64_image}"
                        }
                    }
                ]
            }
        ],
        max_tokens=1000
    )

    return response.choices[0].message.content.strip()
```

```python
def transfer_output(english_solution):
  prompt= f"""You are a professional math translator.

Your task is to translate the following math solution from English to Vietnamese, preserving the original mathematical notations and any LaTeX expressions.

Keep all math expressions, equations, and LaTeX syntax exactly as they are. Only translate the surrounding explanatory text and logical reasoning into natural, fluent Vietnamese.

### Math Solution in English:
{english_solution}

### Translated Solution in Vietnamese:
"""
  messages = [
      {"role": "system", "content": "You are a professional math translator."},
      {"role": "user", "content": prompt}
  ]

  response = client.chat.completions.create(
      model="gpt-3.5-turbo",  # or "gpt-4"
      messages=messages,
      temperature=0.0,  # for deterministic judgment
  )

  result = response.choices[0].message.content.strip()

  return result
```

```python
def rerank(math_problem, responses_list):
    # Tạo prompt yêu cầu GPT-4o chọn response tốt nhất
    prompt = f"""
You are an expert math assistant.

Here is a math problem (in English after translation from Vietnamese):

"{math_problem}"

Below are multiple English translations (with LaTeX formatting) of the problem. Your task:
1. Read all responses carefully.
2. Select the **best** response that is most accurate, clear, formal, and correct (especially LaTeX).
3. Reply with the **index** (starting from 0) of the best response only — do NOT explain anything.

Responses:
"""

    for i, response in enumerate(responses_list):
        prompt += f"\n[{i}]: {response}\n"

    # Gửi đến GPT-4o
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "You are an expert math assistant."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.0,
        max_tokens=10
    )

    result = response.choices[0].message.content.strip()

    try:
        index = int(result)
        return responses_list[index]
    except:
        raise ValueError(f"Invalid response from GPT: {result}")
```

```python
import re

def remove_latex(text):
    # Xóa $$...$$ và $...$
    text = re.sub(r'\$\$(.*?)\$\$', r'\1', text, flags=re.DOTALL)  # công thức block
    text = re.sub(r'\$(.*?)\$', r'\1', text)  # công thức inline

    # Thay \boxed{...} bằng nội dung bên trong
    text = re.sub(r'\\boxed{(.*?)}', r'\1', text)

    # Loại bỏ các ký tự backslash dùng trong LaTeX
    text = text.replace(r"\\", "")
    text = text.replace(r"\cdot", "*")
    text = text.replace(r"\pm", "±")
    text = text.replace(r"\sqrt", "sqrt")
    text = text.replace(r"\frac", "/")  # Hoặc viết hàm cụ thể để parse biểu thức dạng \frac{a}{b}

    return text
```

```python
def solver(origin_prompt):
  origin_prompt = transfer_prompt(origin_prompt)

  input1 = tokenizer1([prompt.format(origin_prompt)], return_tensors="pt").to("cuda")
  input2 = tokenizer2([prompt.format(origin_prompt)], return_tensors="pt").to("cuda")
  input3 = tokenizer3([prompt.format(origin_prompt)], return_tensors="pt").to("cuda")
  input4 = tokenizer4([prompt.format(origin_prompt)], return_tensors="pt").to("cuda")

  output1 = model1.generate(**input1, max_new_tokens = 1024, num_return_sequences=1, do_sample=True, use_cache = True)
  output2 = model2.generate(**input2, max_new_tokens = 1024, num_return_sequences=1, do_sample=True, use_cache = True)
  output3 = model3.generate(**input3, max_new_tokens = 1024, num_return_sequences=1, do_sample=True, use_cache = True)
  output4 = model4.generate(**input4, max_new_tokens = 1024, num_return_sequences=1, do_sample=True, use_cache = True)

  response1 = tokenizer1.batch_decode(output1, skip_special_tokens=True)
  response2 = tokenizer2.batch_decode(output2, skip_special_tokens=True)
  response3 = tokenizer3.batch_decode(output3, skip_special_tokens=True)
  response4 = tokenizer4.batch_decode(output4, skip_special_tokens=True)

  responses = []

  responses.append(response1)
  responses.append(response2)
  responses.append(response3)
  responses.append(response4)

  right_responses = []

  for i, response in enumerate(responses, 1):
      response = response[0]  # Lấy string đầu tiên từ list
      response = response.split('<|im_start|>assistant', 1)[1]

      right_responses.append(evaluate_solution(origin_prompt, response))


  best_response = rerank(origin_prompt, right_responses)

  markdown_output = ""

  best_response = transfer_output(best_response)
  best_response = remove_latex(best_response)
  markdown_output += f"🧠 Phản hồi :\n\n{best_response.strip()}\n\n---\n"

  return markdown_output
```

```python
def load_to_json(data_path):

    knowledge_base = {}

    # Đọc file key-value
    with open(data_path, 'r', encoding='utf-8') as f:
        for line in f:
            if '-' in line:
                key, value = line.strip().split('-', 1)
                knowledge_base[key.strip()] = value.strip()

    return knowledge_base
```

```python
knowledge_base = load_to_json(DEFINITIONS_DATA_PATH)

keys = list(knowledge_base.keys())

# Tự động tạo các câu hỏi với mỗi key
question_templates = [
    "what is {}",
    "define {}",
    "tell me {}",
    "show me {}",
    "meaning of {}",
    "describe {}"
]

training_corpus = []
for key in keys:
    for template in question_templates:
        training_corpus.append(template.format(key))

# Fit vectorizer với toàn bộ keys + câu hỏi mẫu
vectorizer = TfidfVectorizer().fit(keys + training_corpus)
key_vectors = vectorizer.transform(keys)

# Retrieval function
def retrieve_concept(query, knowledge_base, threshold=0.2):
    query_vector = vectorizer.transform([query])
    similarities = cosine_similarity(query_vector, key_vectors).flatten()
    best_idx = similarities.argmax()
    best_score = similarities[best_idx]

    if best_score >= threshold:
        key = keys[best_idx]
        return key, knowledge_base[key]
    else:
        return None, "Xin lỗi, tôi chưa biết khái niệm này."
```

## IV. Chạy chương trình

```python
def mathsolver(text_input, image_input):
    result = ""

    if image_input:
        # OCR + Translation via GPT-4 Vision (this is a placeholder)
        prompt = read_from_image(image_input)
        result += solver(prompt.strip())

    elif text_input.strip():
        # Text translation (your mathsolver function)
        result += solver(text_input)

    elif not text_input.strip() and not image_input:
        result = "⚠️ Please enter text or upload an image."

    return result


def concept_lookup(query):
    # Giả lập tìm khái niệm
    query = GoogleTranslator(source='vi', target='en').translate(query)

    key, response = retrieve_concept(query, knowledge_base)

    #response = GoogleTranslator(source='en', target='vi').translate(response)
    response = transfer_output(response)
    return f"[{key}]: {response}"

# Interface 1: Giải toán
solver_interface = gr.Interface(
    fn=mathsolver,
    inputs=[
        gr.Textbox(label="Nhập bài toán"),
        gr.Image(type="filepath", label="Tải ảnh bài toán")
    ],
    outputs="text",
    title="Math Solver",
    description="Nhập bài toán hoặc tải ảnh để giải toán."
)

# Interface 2: Tra cứu khái niệm
concept_interface = gr.Interface(
    fn=concept_lookup,
    inputs=gr.Textbox(label="Nhập khái niệm"),
    outputs="text",
    title="Math Concept Lookup",
    description="Tra cứu định nghĩa khái niệm toán học."
)

# Gộp 2 interface vào tab
demo = gr.TabbedInterface(
    [solver_interface, concept_interface],
    ["Giải Toán", "Tra Cứu Khái Niệm"]
)

demo.launch(share=True, debug=True)
```
